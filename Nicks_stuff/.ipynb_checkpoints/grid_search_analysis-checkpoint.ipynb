```python
'''
Analyses should be done separately for each comparison

1.) Check n_features X train_trees settings heatmap
'''
S3_2_to_S1_idc = [i for i in grid_search_df.index if 'TableS3.2_to_TableS1' in i]
sns.heatmap(grid_search_df.loc[S3_2_to_S1_idc])

grid_per_comparison = {comparison: [i for i in grid_search_df.index if comparison in i] for comparison in comparisons}
```

```python
source, _, target, _, _, n_features, _, _, train_trees, _, _, best_trees = S3_2_to_S1_idc[0].split('_')
```

```python
sns.heatmap(grid_search_df.loc[grid_per_comparison['TableS3.2_to_TableS1']])
```

```python
'''
Rows: n_features
Cols: train_trees and best_trees
'''
S3_2_to_S1_heatmap_df = pd.DataFrame()

for idx in S3_2_to_S1_idc:
    source, _, target, _, _, n_features, _, _, train_trees, _, _, best_trees = idx.split('_')
    RMSE = grid_search_df.loc[idx, 'RMSE']
    S3_2_to_S1_heatmap_df.loc[f'n_features_{n_features}',f'train_trees_{train_trees}_best_trees_{best_trees}'] = RMSE
```

```python
plt.figure(figsize=(15, 3))
sns.heatmap(S3_2_to_S1_heatmap_df)
```
